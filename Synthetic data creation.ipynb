```python
import numpy as np
from sklearn.linear_model import LogisticRegression
import pandas as pd
```

```python
# Coefficients (replace with your actual coefficients)
coefficients = [0.5, -1.2, 0.8]

# Intercept (replace with your actual intercept)
intercept = -1.5

def logistic_regression_equation(x, coefficients, intercept):
    z = intercept
    for i in range(len(coefficients)):
        z += coefficients[i] * x[:, i]
    return 1 / (1 + np.exp(-z))
```

```python
# Number of data points
n_samples = 1000

# Number of independent variables
n_features = len(coefficients)

# Generate random independent variables
X = np.random.rand(n_samples, n_features)

# Calculate probabilities using the logistic regression equation
y_prob = logistic_regression_equation(X, coefficients, intercept)

# Convert probabilities to binary labels
y = (y_prob >= 0.5).astype(int)
```

```python
df = pd.DataFrame(X)
```

```python
df['y'] = y
```
